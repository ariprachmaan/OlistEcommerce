## Pertumbuhan pemesanan produk dari 10 produk yang paling diminati pada tahun 2018

```python
# Mengolah data tanggal
df_orders["order_purchase_timestamp"] = pd.to_datetime(df_orders["order_purchase_timestamp"], yearfirst=True)
df_orders["month_purchased"] = df_orders["order_purchase_timestamp"].dt.month
df_orders["year_purchased"] = df_orders["order_purchase_timestamp"].dt.year

df_orders = df_orders.loc[df_orders["year_purchased"] == 2017]
df_orders
```

```python
# Menggabungkan tabel order dengan tabel gabungan di objektif 1
df_3 = pd.merge(df1, df_orders, on="order_id", how="inner")

# Menggabungkan dengan data kategori produk dengan pemesanan terbanyak dari objektif 2
df2_top10 = df_2.head(10)
df_3 = df_3.rename({"product_category_name_english":"product_category"}, axis=1).reset_index()
df_3 = pd.merge(df2_top10[["product_category"]], df_3[["product_category",
        "order_item_id", "month_purchased"]], on="product_category", how="left")

df_3
```

```python
# Membuat pivot table
df_3 = pd.pivot_table(df_3.reset_index(), columns="product_category", index="month_purchased", values="order_item_id", aggfunc="sum")
df_3
```

```python
# Membuat visualisasi produk yang paling diminati tahun 2017
ax = df_3.plot(kind="line", title="Pertumbuhan Pemesanan 10 Produk yang Paling Diminati Tahun 2017",
              xlabel="Bulan", ylabel="Banyak Order", figsize=(15,10))

ax.legend(title="Product Category")

for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)
```
